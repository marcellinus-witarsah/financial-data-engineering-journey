```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

```python
def plot_asset_price(df, asset_name, currency):
    fig = go.Figure(
        data=[
            go.Candlestick(
                x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close']
            )
        ],
    )
    fig.update_layout(title=f'{asset_name} Candlestick Chart', xaxis_title='Date', yaxis_title=f'Price ({currency})')
    fig.show()
```

# Financial Market Stock & Cryptocurrency Data

```python
# User input
ticker = input("Enter the ticker symbol (e.g., AAPL for Apple, BTC-USD for Bitcoin): ")
interval = input("Enter the interval (e.g., 1d, 1wk, 1mo): ")

# Get data and visualize it
data = yf.Ticker(ticker)
currency = data.info["financialCurrency"]
df = data.history(period="10y", interval=interval)
plot_asset_price(df, ticker, currency)
```
